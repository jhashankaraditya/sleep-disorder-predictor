## Model Training

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Splitting the data into features and target variable
X = sleep_df.drop('Sleep Disorder', axis=1)
y = sleep_df['Sleep Disorder']
```

```python
X
```

```python
y
```

```python
classifiers = {
    'Logistic Regression': LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=200),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Gaussian Naive Bayes': GaussianNB(),
    'Multinomial Naive Bayes': MultinomialNB(),
    'Bernoulli Naive Bayes': BernoulliNB(),
    'AdaBoost': AdaBoostClassifier()
}
```

```python
# Evaluating each classifier using cross-validation
for name, clf in classifiers.items():
    cv_scores = cross_val_score(clf, X, y, cv=5)  # 5-fold cross-validation
    print(f"{name} Average Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
    print('-' * 60)
```
